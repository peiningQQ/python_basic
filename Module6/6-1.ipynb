```python
# 類別名稱
class Car:
    wheels = 4 # 類別屬性 (Class Attribute)

    #建構函式 -> magic method
    def __init__(self, brand, model, doors=4):
        # 實例屬性
        self.brand = brand
        self.model = model
        self.doors = doors 


# 方法 (Method)
    def car_info(self):
        print(f"Info: Brand: {self.brand} Model:{self.model} has {self.doors} doors")


print(Car.wheels) # 使用 Car 的 類別屬性

car1 = Car("Benz", "C300") # 建立實例
car2 = Car("Toyota", "Altis",5) # 建立實例

print(car1.model)  # car1 的實例屬性
print(car1.wheels)  # car1 的實例屬性
print(car2.model)


car1.car_info()
car2.car_info()
```
