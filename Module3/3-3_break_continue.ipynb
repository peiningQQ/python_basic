##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''
data = "contents"

for char in "contents":
    

    if char == "n":    
     break #立即結束迴圈 不論剩幾次
    print(char, end=",")
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
data = "contents"

for char in "contents":
    

    if char == "o":    
       continue # 跳過continue 以下的程式碼
    print(char, end=",")
```

計算 1 到 100 之間的和，如果值超過500就停止執行

```python
total = 0
for i in range(1,101):
    total += i
    if total > 500:
     break


print(total)

#ipynb 檔案卡住完全沒反應
# Ctrl + Shift + P -> Reload Window
 
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數

```python

total = 0
for i in range(1,101):
    
    if i % 2 != 0 and i % 5 != 0:
     total += i
print(total)
```

```python
total = 0
for i in range(1, 101, 2):
    if i % 5 !=0:
        total += i

print(total)
```
