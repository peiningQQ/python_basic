| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python


a = 4
n= 5

for i in range(n):
    if a % 2 != 0:
     print(a+i*2+1)
    else: 
     print(a+i*2)
```

```python
a = 5
n = 3

if a % 2 != 0:
    a += 1

for i in range(n):
    print(a+i*2)


```

```python
n = 5 # 執行次數
for i in range (n):
    print(" "*(n-i-1) + "* "*(i+1))
```

```python
n = 5 # 執行次數
for i in range (1, n+1):
    print(" "*(n-i) + i*"* ")
```

```python
n = 5
for i in range (n):
    print(" "*i + "* "*(n-i))
```

##### while

```python
#當條件符合時 就一直執行該區塊的程式碼
count = 1

while count <= 5: #結束的條件
    print(count)
    count += 1 #結束的條件
```

```python
count = 1

while True:  #就一直執行底下的程式
    if count > 5: #結束的條件
     break
    print(count)
    count += 1
```
