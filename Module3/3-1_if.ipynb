條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
fruit = "Apple"

if fruit == "Apple":
    print("fruit is Apple")


```

##### if...else...

```python
fruit = "Banana"

if fruit == "Apple":
    print("fruit is Apple")
else: #當if 語句沒有達成 就執行else區塊的程式碼
    print("fruit is not Apple")
```

```python
num = 65

if num > 60:
    result =  "及格"

else:
    result =  "不及格"

#= 三元運算
result =  "及格" if num > 60 else "不及格"
print(result)




```

##### if...elif... (else)

```python
fruit = "Banana"


if fruit == "Apple":
    print("fruit is Apple") #執行另一段程式碼
elif fruit == "Banana":
    print("fruit is Banana")

elif fruit == "Apple":#當前面條件已符合 後面不判斷 
    print("fruit is not Apple")
else: #當if 語句沒有達成 就執行else區塊的程式碼
    print("fruit is not Apple")
```
