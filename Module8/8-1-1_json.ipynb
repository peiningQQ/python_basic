### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
|`json.dump(<obj>, <file>)` |物件轉JSON並寫入文件|
|`json.dumps(<obj>)` |物件轉JSON字串|
|`json.load(f)` |讀取JSON文件|
|`json.loads(f)` |將JSON字串轉成Python物件|

```python
import json
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}

with open('result.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(fruit_dict, indent=4))
```

```python
import json
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}

with open('result.json', 'w', encoding='utf-8') as f:
    json.dump(fruit_dict, f, indent=4)
```

```python
with open('result.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

print(data)
print(type(data))
print(data['apple'])
```

```python
with open('result.json', 'r', encoding='utf-8') as f:
    str_data = f.read()

    print(str_data)
    print(type(str_data))

data = json.loads(str_data) # 來源不一定是從檔案 很常需要這樣做轉換
print(type(data))
print(data)
print(data[0])
```
