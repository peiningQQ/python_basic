##### 讀取檔案 read( )

```python
# 印出 50 個 "="
print("="*50)

# 讀取檔案內容 - 常用的寫法 (執行完畢會自動關閉檔案)
# 相對路徑
with open(file="../Module2/2-1.py", mode="r", encoding="utf-8") as f:
    texts = f.read()

print(texts)

# 印出 50 個 "="
print("="*50)
```

```python
# 一行一行讀取
with open(file="../Module2/2-1.py", mode="r", encoding="utf-8") as f:
    for line in f:
        print(line)
```

```python
# 一行一行讀取
with open(file="../Module2/2-1.py", mode="r", encoding="utf-8") as f:
    text_list = f.readlines()

print(text_list)
```

```python
# 資料處理
with open(file="../Module2/2-1.py", mode="r", encoding="utf-8") as f:
    text_list = f.readlines()

result = []
while '\n' in text_list:
    text_list.remove('\n')

for text in text_list:
    text_without_n = text.rstrip('\n')

    result.append(text_without_n)

print(result)
```

##### 寫入檔案 write( )

```python
# 文字內容
text = '''
傑出的人會努力規劃自己的人生，普通的人則是隨便命運怎麼安排
一個是充實、有目的的過每一天，一個是隨興的過一天算一天
創業這條路 Michael E.Gerber
'''

# 寫入檔案
with open('test.txt', 'w', encoding='utf-8') as f:
    f.write(text)
```

```python
# 寫入資料到最後一行 (附加文字在檔案內容最後一行)
text = "People say that success just happens to you. It doesn't!"
with open('test.txt', 'a', encoding='utf-8') as f:
    f.write(text+'\n')
```

```python
nums = [i for i in range(1, 101) if i % 2 == 0]
print(nums)

with open('result.txt', 'w') as f:
    f.write(str(nums))
```
