```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
friut_info = {
    "apple" : {
    "friut_price" : 100,
    "amount" : 50,
    "color" :[ "green", "red"],

    },

    "orange" : {
    "friut_price" : 200,
    "amount" : 30,
    "color" : ["orange"],

    },

    "tomota" : {
        "friut_price" : 250,
        "amount" : 10,
        "color" :["red"],

        }

}

data = {

    "banana" : {
            "friut_price" : 30,
            "amount" : 100,
            "color" :["yellow"],

            }

}




print(friut_info)
```

```python
# 從 Module 引入 function

from pprint import pprint
pprint(friut_info)
```

```python
# dict 的合併
from pprint import pprint
friut_info.update(data)
pprint(friut_info)

```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色

'''
new_color = "yellow"

if "yellow" not in friut_info["apple"]["color"]: #避免重複新增
    friut_info['apple']['color'].append("yellow") # 新增顏色

# 假設有不小心多新增 color清除多餘的
# while friut_info['apple']['color'].count(new_color)>1:
#     friut_info['apple']['color'].remove(new_color)


pprint(friut_info)

friut_info['tomota']['color'].append("green")
pprint(friut_info)
```
