```python
# f-string (又稱作 formatted string literals, version >= 3.6)

name = "Penny"
age = 20

msg = f"{name} is { age} years old."
print(msg)
```

```python
# 靠左對齊、靠右對齊
msg = f"Name: [{name:10}], Age:[{age:>10}]"
print(msg)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)
msg = f"Name: [{name:X<10}], Age:[{age:X>10}]"
print(msg)
```

```python
# 置中對齊
msg = f"Name: [{name:^10}], Age:[{age:^10}]"
print(msg)
```

```python
# 小數點輸出幾位

result = 3.1256722
print(f"輸出小數點2位: {result:.2f}")


```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = f"Name: [{name:.2}], Age:[{age:^10}]"
print(msg)

```

```python
total = 12248982222222
print(f"{total:,}")


```
