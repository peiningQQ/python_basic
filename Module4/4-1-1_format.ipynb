```python
# 多組文字

msg = "%s,%s!" % ("Hello","World")
print(msg)
```

```python
# 整數
msg = "%d 元" % (300)
print(msg)

```

```python
# 文字與整數
msg = "%s is %d years old." % ("Penny",30)
print(msg)

```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = "[%10s]" % ("Hello")
print(msg)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)

msg = "[%-10s]" % ("Hello")
print(msg)
```

```python
# 指定浮點數位數

msg = "[%.2f]" % (3.14567)
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)

msg = "[%.3s]" % ("Hello")
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
sql = "A123456789"
msg = "[%.3s*****]" % (sql)
print(msg)
```

```python
# 空白補 0

msg = "[%06d]" % (567)
print(msg)
```
