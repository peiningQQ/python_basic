```python
# 字串變數初始化
name = "Alex"
print(name)


name = "Jarvis" #重新赋值
print(name)
```

```python
# 替換字串
# 因為字串不可變 每次修改字串 都需要重新赋值的動作/或是用新的變數去赋值
name = "Alex"
'''
str.replace(<old>,<new>)
將 string 中的 str1 替換成 str2
'''
# ex 替換成 len
new_name = name.replace("ex", "len")# 會回傳(return)新的資料 所以要用變數去接

print(name)
print(new_name)

```

```python
# 去除兩側空格
data = "_Hello!World_"

'''
str.strip()
去除字串 string 左、右兩邊的字元
'''
new_data = data.strip("_")
print(data)
print(new_data)


new_data2 = data.lstrip("_")
print(new_data2)

new_data3 = data.rstrip("_")
print(new_data3)


```

```python
# 字串小寫變大寫

msg = "HELLO WORLD"

'''
str.lower()
將字串 string 裡的字母全部改成小寫
'''
lower_msg = msg.lower()
print(lower_msg)


msg2 = "hellow world"

'''
str.upper()
將字串 string 裡的字母全部改成大寫
'''
upper_msg = msg2.upper()
print(upper_msg)


```

```python
# 字串分割成 list
greeting = "hello world! hello John"

world_list = greeting.split()
print(world_list)


```

```python
greeting = "hello world! hello John"

data = data.replace(" ", "__")
print(data)

world_list = greeting.split("__") #可以指定依照什麼座分隔
print(world_list)
```
