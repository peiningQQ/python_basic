```python
# tuple 初始化: 第一種
class_tuple = ("AIPE", "BDSE")
print(class_tuple)

```

```python
# tuple 初始化: 第二種

class_tuple = "AIPE", "BDSE"
print(class_tuple)

```

```python
a, b = (1,2)
print(a)
print(b)


```

```python
# 透過指定索引輸出值


class_tuple = "AIPE", "BDSE"
print(class_tuple[1])
```

```python
# 複數變數修改值
x, y = 100, 20
print(x, y)
```

```python
# tuple 不可以修改指定索引的值
class_tuple = "AIPE", "BDSE"
class_tuple[0] = "iSpan"
```

```python
# ( )可以修改指定索引的值

class_tuple = "AIPE", "BDSE"
data = list(class_tuple)
data[0] = "iSpan"
print(data)
```

```python
# 用 for 迴圈逐一輸出資料
class_tuple = "AIPE", "BDSE"
for i in class_tuple:   
    print(i)
```

```python
# 用 len 計算 tuple 資料個數

class_tuple = "AIPE", "BDSE"
print(len(class_tuple))
```
