##### slice 透過 [ : ] 進行分割
`[n:m]` 走訪 n 到 m - 1 的元素

```python
# 初始化一個字串
data = "Hello,World!"
print(data)
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
print(data[0:5])
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
print(data[:5])
```

```python
# 從 6 取到尾
print(data[6:])
```

```python
# 從 7 取到 9（不包含第 9 個元素，實際為 7 ~ 8）
print(data[7:9])
```

##### 使用負號

```python
# 從倒數第 10 取到倒數第 7（不包含倒數第 7 的元素）
print(data[-10:-7])
```

```python
# 從倒數第 6 取到尾
print(data[-6:])
```

```python
# 從頭取到倒數第 7（不包含倒數第 7 的元素）
print(data[:-7])

```

```python
# 也可以使用在 list
data = [1, 2, 3, 4, 5]
print(data[-3:])
```

```python
# 實做練習 將每個單字改成大寫開頭
string04 = "hello world! hello john"
word_list = string04.split(" ")

print(word_list)

result = ""
for word in word_list:
    result += word[0].upper() + word[1:] + " "
print(result)

# string04.capitalize
```

```python
# 實做練習 將每個單字改成大寫開頭
string04 = "hello world! hello john"
word_list = string04.split(" ")

result = []
for word in word_list:
    # print(word)
    result.append(word[0].upper() + word[1:])

print(result)

ans = " ".join(result)
print(ans)
```
