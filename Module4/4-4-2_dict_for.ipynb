```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': 'orange'
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1
# 拿出所有的 key 
for key in fruit_dict:
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2 -> dict.value()
# 拿出所有的 value
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 資料3 -> dict.items()
# 同時拿出 key value
for key, value in fruit_dict.items():
    print(key)
    print(value)
    print("="*50)
```

```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}

# 判斷某個 key 是否存在於 dict
new_fruit = "banana"
if new_fruit not in fruit_dict:
    print(f"{new_fruit} 不在 dict 中")
```

```python
# 取得 dict 的 value
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}

print(fruit_dict["apple"])

# Python 表達不存在 -> None (null)
# 另一個方式 dict.get(<key>, {}) 避免 dict 沒有這個 key 就報錯
print(fruit_dict.get("apple", {}))
print(fruit_dict.get("banana", {}))
print(fruit_dict.get("banana"))
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}
]

fruit_dict = {}

for fruit_color in fruit_color_list:
    # print(fruit_color)

    for fruit, color in fruit_color.items():
        # print(fruit, color)
        if fruit_dict.get(fruit) is None: 
            fruit_dict[fruit] = [color]
        else:
            fruit_dict[fruit].append(color)

print(fruit_dict)
```

```python
fruit_dict = {}

for fruit_color in fruit_color_list:
    # print(fruit_color)

    for fruit, color in fruit_color.items():
        # print(fruit, color)
        if fruit not in fruit_dict: 
            fruit_dict[fruit] = [color]
        else:
            fruit_dict[fruit].append(color)

print(fruit_dict)
```

##### `defaultdict(<type>)`

```python
from collections import defaultdict

fruit_dict = defaultdict(int)

print(fruit_dict)

# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
print(fruit_dict['apple'])

# 以int型態的預設值 0 來初始化
print(fruit_dict)
```

```python
from collections import defaultdict

fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}
]

fruit_dict = defaultdict(list) # 呼叫到 key 時 直接建立空的 []

for fruit_color in fruit_color_list:
    for fruit, color in fruit_color.items():
        fruit_dict[fruit].append(color)

print(fruit_dict)

fruit_dict.update({
    "banna": "yellow"
})

fruit_dict = dict(fruit_dict)
print(fruit_dict)
```
