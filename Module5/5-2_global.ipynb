```python
# 全域變數
name = "Penny"

# 定義函式
def greeting():

    name = "Pochita"
    print(f"Hello,{name}")

# 執行函式
greeting()

# 輸出 name
print(name)
```

##### 使用 global 讓函式存取全域變數的值

```python
# 全域變數
name = "Penny"

# 定義函式
def greeting():

    global name
    name = "Pochita"
    print(f"Hello,{name}") #修改全域變數

# 執行函式
greeting()

# 輸出 name
print(name)
```
