```python
# 建立 function 1 沒有參數 沒有 return 值
def greeting(): 
    print("Good Morning!!")

greeting()
x = greeting()
print(x)
```

```python
# str.method 通常都有 return 值
# 建立 function 2 有 return 值
def get_name():
    name = "Jarvis"
    return name

data = get_name()
print(data)
```

```python
# 建立 function 3 帶參數
def greeting(name, greet):
    print(f"Hello {name}, {greet}")

greeting("Pochita", "Good Morning")
greeting("Denji", "Hello World!")

# 沒有預設參數值 在使用時就一定要給該參數
greeting("Denji")
```

```python
# 建立 function 4 預設參數值
def greeting(name, greet="Good Night"):
    print(f"Hello {name}, {greet}")

greeting("Jarvis")
greeting("Jarvis", "Good Morning!")
```

```python
# 改寫 Module2 practice 
costs = [10000, 4800, 6500, 2300, 4900]

def after_discount(total, threshold=2000, discount=200):
    if total <= 0:
        return 0
    
    if total >= threshold:
        # 需要折幾個 200
        total -= discount * (total//threshold)
    
    print(f"折扣後: {total} 元")
    return total

ans = after_discount(-100)
print(ans)
ans = after_discount(100)


# for cost in costs:
#     ans = after_discount(cost, 3000, 300)
#     print(ans)
```
