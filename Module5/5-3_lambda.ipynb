### lambda 匿名函式  
用法  

`<fuction_name> = lambda <arguments>: <expression>`

````python
# 建立函式
def add(x,y):
    return x + y

result = add(100,50)
print(result)
````

````python
# 基本用法
add2 = lambda x, y: x + y
result = add2(30, 100)
print (result)
````

````python
# return index 2 的資料

def get_index_2(data):
     return data[2]

get_index_2 = lambda data: data[2]

result = get_index_2([10, 20, 30,40])
print(result)
````

````python
# 條件判斷

def abs(num):
    if num >=0:
        return num
    else:
        return -num
    
result = abs(-100)
print(result)


````

````python
abs2 = lambda num: num if num>= 0 else -num
result = abs2(-200)
print(result)
````

### 小試身手  
請建立一個 list ，並在該 list 內產生多個 `[1, 2, x]` 的 list (x 的值要遞增)  
x 到 10 即可  
Ex :  
```Python
[ [1, 2, 1], [1, 2, 2], [1, 2, 3]...[1, 2, 10] ]
```

````python
result = [ ]
for i in range(1, 11):
    result.append([1, 2, i])

print(result)

````

````python
result = [[1,2,3] for i in range(1,11)]
print(result)
````

##### `list.sort()` 方法

````python
nums = [10,5,39,3,100]
nums.sort()

print(nums)
print(nums[0]) #最小
print(nums[-1]) #最大


````

````python
result = [[1,2,3] for i in range(1,11)]
print(result)
result.sort(reverse=True) #從大排到小
print(result)
````

每三個 list 中間的值要改成 5  
Ex:  
```python
[ [1, 2, 1], [1, 2, 2], [1, 5, 3], [1, 2, 4], [1, 2, 5], [1, 5, 6]...]
```

````python
result = []
for i in range(1,11):
    if i % 3 == 0:
        result.append([1,5,i])
    else:
        result.append([1,2,i])
print(result)

result.sort(reverse=True)
print(result)

result.sort(reverse=True, key=lambda x: x[2])
print(result)
````
