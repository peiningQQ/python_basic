```python
# 自訂例外處理
def pdf_func(page):
    if type(page) != int:
        raise TypeError("資料行錯誤")
    if page <= 0:
        raise ValueError("Page 不能小於0")
    # 處理 PDF 的程式碼
pdf_func([1, 2, 3])
```

```python
def get_user_data(permission, user_class):
    if permission != "Teacher":
     raise PermissionError("權限錯誤")


    if user_class != "Bigdata":
     raise ValueError("班級錯誤")
    return [ 49, 69, 33, 60,90]
try:
  grades = get_user_data("Teacher", "AIPE")
except PermissionError as e:
  grades = e
except ValueError as e:
  grades = e

print(grades)
```
