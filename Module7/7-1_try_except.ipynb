##### try...except...

```python
# 多層例外處理:
try:
 x = 100
 y = [1,2]


 print(y)
 print(y[2])
 print(x)


except NameError as e: 
  print(f"發生NameError:{e}") 

except IndexError as e: 
  print(f"發生IndexError:{e}") 


except Exception as e: # 當 try 底下發生錯誤 就要執行程式碼
  print(f"程式發生錯誤:{e}")

print("繼續執行")
```
