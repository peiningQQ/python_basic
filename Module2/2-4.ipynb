比較運算子 (用 if 判斷)

```python

a = 3
b = 4

print(a==b) #回傳 False
print(a!=b) #回傳 Ture

if a != b: #if  Ture 較要執行 縮排內的程式碼
    print("a 不等於 b")

print("不管 if 都會執行") # shift + Tab 反縮排
```

邏輯運算子 and / or / not

```python
role = "worker"
age = 20

# 條件都符合 才為Ture
if role != "student" and age > 18:
    print("已滿18歲且不為學生")

# 其中一個條件都符合 就為Ture
if role != "student" or age > 20:
    print("已滿20歲且腳色學生")

# if not False才執行
if  not role == "student":
    print("角色不為學生")
```
