整數 (int)、浮點數 (float)

```python
num1 = 10
num2 = 5.5
print(num1)
print(num2) #函式 
```

布林值 (boolean)

```python
parse_ok = True
parse_not_ok = False
print(parse_ok)
print(parse_not_ok)
```

字串 (string)

```python
p = "True"
print(p)
print(type(p))
```

```python
# type() 檢查資料型別
x = "5"
y = 10

print(x)
print(type(x))
print(y)
print(type(y))

total = x + y
print(total)
```

```python
name = "Penny"
msg ='Hello World'
print(name)
print(msg)
```

列表 (list)

```python
num_list = [1, 2, 3]
print(num_list)
```

元組 (tuple)

```python
class_tuple = ("BOSE", "AIPE", "AIOT")
print(class_tuple)
print(type(class_tuple))
```

集合 (set)

```python
num_set = {1, 2, 3}
print(num_set)
print(type(num_set))

```

字典 (dict)

```python
#{key: key}
user_info ={"name":"Penny","age": 30 }
print(user_info)
print(type(user_info))
```

```python
#創建空的list
data = []
print(data)
print(type(data))
```

```python
# 創建一個空的 dict
data = {}
print(data)
print(type(data))
```

```python
# 創建一個空的 set
data = set()
print(data)
print(type(data))
```
